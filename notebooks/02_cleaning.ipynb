# 02 Cleaning

```python
import pandas as pd
```

```python
shootings = pd.read_csv("../Data/Raw/ShootOut_RawDataset.csv")
agencies = pd.read_csv("../Data/Raw/Agency_RawDataset.csv")
```

### Drop Columns

```python
shootings = shootings.drop(columns=["race_source", "location_precision"], errors="ignore")
agencies = agencies.drop(columns=["oricodes"], errors="ignore")
```

### Type Fixes

```python
shootings['date'] = pd.to_datetime(shootings['date'], errors='coerce')
shootings['age'] = pd.to_numeric(shootings['age'], errors='coerce')
```

### Clean Strings

```python
shootings['gender'] = shootings['gender'].str.lower().str.strip()
shootings['armed_with'] = shootings['armed_with'].fillna("unknown")
```

### Map Race

```python
race_map = {
    "W": "White",
    "B": "Black",
    "H": "Hispanic",
    "A": "Asian",
    "N": "Native American",
    "O": "Other"
}
shootings['race'] = shootings['race'].map(race_map).fillna("Unknown")
```

### Boolean Conversions

```python
shootings['was_mental_illness_related'] = shootings['was_mental_illness_related'].astype(bool)
shootings['body_camera'] = shootings['body_camera'].astype(bool)
```

### Fixing Multiple Agency Ids

```python
shootings['agency_ids'] = shootings['agency_ids'].astype(str)

shootings = shootings.assign(
    agency_id=shootings['agency_ids'].str.split(',')
).explode('agency_id')

shootings['agency_id'] = pd.to_numeric(shootings['agency_id'], errors='coerce')
```

# Saving

```python
shootings.to_csv("../Data/Cleaned/clean_shootings.csv", index=False)
agencies.to_csv("../Data/Cleaned/clean_agencies.csv", index=False)
```
